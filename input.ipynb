```python
#function to validate input
def validate5():
    while True:
        try:
            a=int(input('please enter value between 0 and 5 inclusive: '))
            break
        except Exception as e:
            print('Invalid data')
    while a>5 or a<0:
        print('number outside of range:')
        a=int(input('please enter number between 0 and 5 inclusive: '))
    return a
```

```python
#function to validate yes or no answers(0 or 1)
def validate1():
    while True:
        try:
            a=int(input('please enter value 0 or 1: '))
            break
        except Exception as e:
            print('Invalid data')
    while a>1 or a<0:
        print('number outside of range:')
        a=int(input('please enter value 0 or 1: '))
    return a
```

```python
#bubble sort- to sort the SORT array

def bubble_sort(array):
    last=len(array)-1
    while True:
        swap=False
        for i in range(last):
            if array[2,i]<array[2,i+1]:
                array[i],array[i+1]=array[i+1],array[i]
                categories[i],categories[i+1]=categories[i+1],categories[i]
                swap=True
        if not swap:
            break
        last=last-1
    return array
```
